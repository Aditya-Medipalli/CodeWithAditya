# Coding Part

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4),dpi=200),

sns.boxplot(data=df,y='reading score',x='parental level of education',
           hue='test preparation course',palette='Set2')
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,4),dpi=200),

sns.violinplot(data=df,x='reading score',y='parental level of education')
#plt.legend(bbox_to_anchor=(1.05,0.5))
```

# Seaborn Grids

```python
df.head()
```

```python
#Cat Plot
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',
           col='lunch',row='race/ethnicity')
```

```python
sns.pairplot(df)
```

```python
g = sns.PairGrid(df,hue='gender')

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.kdeplot)

g = g.map_lower(sns.kdeplot)

g = g.add_legend()
```
